```python
holidays=holidays[(holidays['transferred']==False) & (holidays['type']!='Work Day')]
holidays.drop(['transferred','description'],axis=1,inplace=True)
holidays.loc[holidays['locale']=='Regional','state'] = holidays['locale_name']
holidays.loc[holidays['locale']=='Local','city'] = holidays['locale_name']
holidays.loc[holidays['locale']=='National','National_holiday'] = 1
holidays.loc[holidays['locale']=='Regional', 'Regional_holiday'] = 1
holidays.loc[holidays['locale']=='Local', 'Local_holiday'] = 1
holidays.fillna({'National_holiday':0,'Regional_holiday':0,'Local_holiday':0},inplace=True)
holidays = holidays.groupby(['date','locale_name']).first()
holidays.reset_index(inplace=True)
holidays[['National_holiday','Regional_holiday','Local_holiday']] = holidays[['National_holiday','Regional_holiday','Local_holiday']].astype('uint')
# holidays
```

```python
def dty_sincos(a):
    a['date']=pd.to_datetime(a['date'])
    (date, month, year) = (a['date'].dt.day, a['date'].dt.month, a['date'].dt.year)
    DMY = pd.DataFrame({
        'date': date.values,
        'month': month.values,
        'year': year.values
    })
    (DMY['date_sin'],DMY['date_cos'],DMY['month_sin'],DMY['month_cos'],DMY['year']) = (np.sin(2*np.pi*DMY['date']/31), np.cos(2*np.pi*DMY['date']/31),np.sin(2*np.pi*DMY['month']/12),np.cos(2*np.pi*DMY['month']/12),DMY['year'] - DMY['year'].min())
    DMY.drop(['date','month'], axis=1, inplace=True)
    a = pd.concat([a, DMY],axis=1)
    return a
```

```python
def dat_prep(a):
#     a = pd.merge(a,transactions,how='left',on=['date','store_nbr'])
    a = pd.merge(a,oil,how='left',on='date')
    a = pd.merge(a, stores,how = 'left',on='store_nbr')
    a = a.merge(holidays[holidays['locale']=='Local'][['date','city','Local_holiday']],on=['date','city'],how='left')
    a = a.merge(holidays[holidays['locale']=='Regional'][['date','state','Regional_holiday']],on=['date','state'],how='left')
    a = a.merge(holidays[holidays['locale']=='National'][['date','National_holiday']],on=['date'],how='left')
    a = dty_sincos(a)
    a['store_nbr'] = a['store_nbr'].astype(object)
    a.fillna({'National_holiday':0,'Regional_holiday':0,'Local_holiday':0},inplace=True)
#     a['transactions']=a['transactions'].interpolate(method='linear',limit_direction='both')
    a['dcoilwtico'] = a['dcoilwtico'].interpolate(method='linear',limit_direction='both')
#     a = pd.concat([a,pd.get_dummies(a['type'],prefix='type',drop_first=True)],axis=1)
    a[['National_holiday','Regional_holiday','Local_holiday']] = a[['National_holiday','Regional_holiday','Local_holiday']].astype('uint')
#     a.replace(to_replace=['HOME AND KITCHEN I','HOME AND KITCHEN II','HOME APPLIANCES','HARDWARE','PLAYERS AND ELECTRONICS','HOME CARE','CLEANING','LAWN AND GARDEN'],value=1,inplace=True)
#     a.replace(to_replace=['BABY CARE','PERSONAL CARE','BEAUTY','LADIESWEAR','LINGERIE'],value=2,inplace=True)
#     a.replace(to_replace=['BEVERAGES','BREAD/BAKERY','CELEBRATION','DAIRY','DELI','EGGS','FROZEN FOODS','GROCERY I','GROCERY II','LIQUOR,WINE,BEER','MEATS','PREPARED FOODS','PRODUCE','SEAFOOD','POULTRY','PET SUPPLIES'],value=3,inplace=True)
#     a.replace(to_replace=['SCHOOL AND OFFICE SUPPLIES','MAGAZINES','BOOKS'],value=4,inplace=True)
#     a.replace(to_replace=['AUTOMOTIVE'],value=5,inplace=True)
#     a = pd.concat([a,pd.get_dummies(a['family'],prefix='family',drop_first=True)],axis=1)
#     a.drop('type',axis=1,inplace=True)
    return a
```
